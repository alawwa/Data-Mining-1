```python
#List
fruits = []
fruits = ["apple", "banana"]
another_list = ["this is string", 10] #list dengan type yang berbeda

print (fruits)
print (another_list)

combined = [fruits, another_list] #menggabungkan
print (combined)
```

```python
#Menambahkan elemen kedalam list dapat menggunakan append, tanda + atau combined
fruits = ["apple", "banana"]
fruits.append("orange")
print (fruits)

fruits = fruits +["mango"]
print (fruits)

fruits = ["apple", "banana"]
another_fruits = ["mango"]
combined = [fruits, another_fruits]
print(combined)
```

```python
#Data Integrasi
import pandas as pd #wajib diinstal
import numpy as np
```

```python
##Merging
left = pd.DataFrame({'key':['K0','K1','K2','K3'],
                    'A': ['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3']})

right = pd.DataFrame({'key':['K0','K1','K2','K3'],
                    'C': ['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})

result1 = pd.merge(left,right, on='key')
```

```python
result1
```

```python
left = pd.DataFrame({'key':['K0','K1','K2','K3'],
                    'A': ['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3'],
                    'C': ['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})

right = pd.DataFrame({'key':['K0','K1','K2','K3'],
                    'E': ['E0','E1','E2','E3'],
                    'F':['F0','F1','F2','F3']})

result1 = pd.merge(left,right, on='key')
```

```python
result1
```

```python
#Multiple Key
left = pd.DataFrame({'key1':['K0','K0','K1','K2'],
                     'key2':['K0','K1','K0','K1'],
                    'A': ['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3']})

right = pd.DataFrame({'key1':['K0','K1','K1','K2'],
                     'key2':['K0','K0','K0','K0'],
                    'C': ['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})

result2 = pd.merge(left,right, on=["key1","key2"])
result2 #yang berhasil digabungkan hanya yang memiliki key1 dan key2 sama, kalau ada double maka akan dilist satu per satu
```
