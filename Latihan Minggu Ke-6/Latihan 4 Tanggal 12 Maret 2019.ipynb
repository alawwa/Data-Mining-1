```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('Latihan4.csv')
data.head()
```

```python
data.describe()
```

```python
np.sum(data.isnull())
```

```python
data['MinTemp']=data['MinTemp'].fillna((data['MinTemp'].median()))
data['MaxTemp']=data['MaxTemp'].fillna((data['MaxTemp'].median()))
data['Rainfall']=data['Rainfall'].fillna((data['Rainfall'].median()))
data['Evaporation']=data['Evaporation'].fillna((data['Evaporation'].median()))
data['Sunshine']=data['Sunshine'].fillna((data['Sunshine'].median()))
data['WindGustSpeed']=data['WindGustSpeed'].fillna((data['WindGustSpeed'].median()))
data['WindSpeed9am']=data['WindSpeed9am'].fillna((data['WindSpeed9am'].median()))
data['WindSpeed3pm']=data['WindSpeed3pm'].fillna((data['WindSpeed3pm'].median()))
data['Humidity9am']=data['Humidity9am'].fillna((data['Humidity9am'].median()))
data['Humidity3pm']=data['Humidity3pm'].fillna((data['Humidity3pm'].median()))
data['Pressure9am']=data['Pressure9am'].fillna((data['Pressure9am'].median()))
data['Pressure3pm']=data['Pressure3pm'].fillna((data['Pressure3pm'].median()))
data['Cloud9am']=data['Cloud9am'].fillna((data['Cloud9am'].median()))
data['Cloud3pm']=data['Cloud3pm'].fillna((data['Cloud3pm'].median()))
data['Temp9am']=data['Temp9am'].fillna((data['Temp9am'].median()))
data['Temp3pm']=data['Temp3pm'].fillna((data['Temp3pm'].median()))
data['RISK_MM']=data['RISK_MM'].fillna((data['RISK_MM'].median()))
np.sum(data.isnull())
```

```python
data['WindGustDir']=data['WindGustDir'].fillna(data['WindGustDir']).dropna().mode()[0]
data['WindDir9am']=data['WindDir9am'].fillna(data['WindDir9am']).dropna().mode()[0]
data['WindDir3pm']=data['WindDir3pm'].fillna(data['WindDir3pm']).dropna().mode()[0]
data['RainToday']=data['RainToday'].fillna(data['RainToday']).dropna().mode()[0]
data['RainTomorrow']=data['RainTomorrow'].fillna(data['RainTomorrow']).dropna().mode()[0]
data.isnull().sum()
```

```python
#mengatasi outlier
from scipy import stats
z=np.abs(stats.zscore(data._get_numeric_data()))
print(z)
data1=data[(z<3).all(axis=1)]
data1.head()
print(data1.shape)
```

```python
data1
```

```python
#data untuk transformasi
data2=data1.drop(['Date','Location','WindGustDir','WindDir9am','WindDir3pm','RainToday','RainTomorrow'],1)
data2
```

```python
xx = data2.iloc[0:134646,0:17]
xx.head()
```

```python
#normalization
from sklearn.preprocessing import normalize
xnorm=normalize(xx,norm='l2') #default norm L2, other are L1 and max
xnorm=pd.DataFrame(xnorm) #biar bentuknya tabel
xnorm.tail()
```

```python
#minmax
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(-5,1)) #bebas nilainya yang diinginkan
scaler.fit(xx)
x_scaled= scaler.transform(xx)
x_scaled=pd.DataFrame(x_scaled)
x_scaled.tail()
```

```python
#3sigmoid
xsigmoidal=(1-2.7181823**(-xnorm))/(1+2.7181823**(-xnorm))
xsigmoidal=pd.DataFrame(xsigmoidal)
xsigmoidal.tail()
```

```python
#decimal scalling
xdecimal=xx/(10**3) #angka 3 adalah bebas
xdecimal=pd.DataFrame(xdecimal)
xdecimal.tail()
```

```python
#softmax, mengubah nilai dari 0 sampai 1 untuk transformasi
def softmax(xx):
    return np.exp(xx) / np.sum(np.exp(xx), axis=0)
xsoft=softmax(xnorm)
xsoft=pd.DataFrame(xsoft)
print(xsoft)
```

```python
#feature selection (mencari skor tertinggi lebih dulu)
from sklearn.feature_selection import SelectKBest, chi2, f_regression
Xx=xx.iloc[0:134646,0:17]
y=xx.iloc[0:134646,1]
selector=SelectKBest(f_regression,k="all").fit(Xx,y)
X_new=selector.transform(Xx)
scores=selector.scores_
scores
```
