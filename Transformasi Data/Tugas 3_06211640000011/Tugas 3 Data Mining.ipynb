## Preprocessing Data dan Regresi untuk Mendapatkan Model Terbaik pada Data Breast Cancer Combria

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('Tugas3.csv')
data.head()
```

```python
data.describe()
```

```python
np.sum(data.isnull())
```

### A. Mengatasi Missing Value Data Breast Cancer Combria

```python
data['BMI']=data['BMI'].fillna((data['BMI'].median()))
data['Leptin']=data['Leptin'].fillna((data['Leptin'].median()))
np.sum(data.isnull())
```

```python
datanomissing = data
datanomissing.head()
```

```python
datanomissing.to_excel('Tugas 3a.xlsx')
```

### B. Mengatasi Missing Value dan Outlier Data Breast Cancer Combria

```python
from scipy import stats
z = np.abs(stats.zscore(datanomissing._get_numeric_data()))
print(z)
datanomissingoutlier=data[(z<3).all(axis=1)]
print(datanomissingoutlier.shape)
```

```python
datanomissoutlier.head(10)
```

```python
datanomissingoutlier.to_excel('Tugas 3b.xlsx')
```

### C. Mengatasi Missing Value dan Melakukan Transformasi Data Breast Cancer Combria 

```python
x = data.iloc[0:106,0:6]
x.head()
```

```python
#normalization
from sklearn.preprocessing import normalize
xnorm = normalize(x,norm='l2') #default norm L2 and max
xnorm = pd.DataFrame(xnorm)
xnorm.tail()
```

```python
def softmax(x):
    return np.exp(x)/np.sum(np.exp(x), axis = 0)
xsoft = softmax(xnorm)
xsoft = pd.DataFrame(xsoft)
print(xsoft)
```

```python
xsoft.to_excel('Tugas 3c.xlsx')
```

### D. Mengatasi Missing Value, Outlier, dan Melakukan Transformasi Data Breast Cancer Combria

```python
xx = datanomissingoutlier.iloc[0:106,0:6]
xx.head()
```

```python
from sklearn.preprocessing import normalize
xxnorm = normalize(xx,norm='l2') #default norm L2 and max
xxnorm = pd.DataFrame(xxnorm)
xxnorm.tail()
```

```python
def softmax(xx):
    return np.exp(xx)/np.sum(np.exp(xx), axis = 0)
xxsoft = softmax(xxnorm)
xxsoft = pd.DataFrame(xxsoft)
print(xxsoft)
```

```python
xxsoft.to_excel('Tugas 3d.xlsx')
```
